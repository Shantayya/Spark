```python
from bs4 import BeautifulSoup
import requests
url = 'https://quotes.toscrape.com/'
r = requests.get(url)
data = r.text
soup = BeautifulSoup(data,'html.parser')
print(soup)
```

```python
print(soup.span)
```

```python
print(soup.title)
```

```python
print(soup.title.parent)
```

```python
print(soup.title.string)
```

```python
print(soup.span.string)
```

```python
for tag in soup.find_all('span'):
    print(tag)
```

```python
for tag in soup.find_all('span',{'class':'text'}):         #in find_all(tag,attribute key value) here span is tag and class(key)="text"(value) is attribute
    print(tag.string)
    print('-----------------------------')
```

```python
#extract author names using bs4
import requests
from bs4 import BeautifulSoup
url = 'https://quotes.toscrape.com/'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
for data in soup.find_all('small',{'class':'author'}):
    print(data.string)
```

```python
#accessing attributes
print(soup.span.attrs)
```

```python
sample = '<b><span class="abc xyz"></span></b>'   #span tag has multivalues attributes like class=['abc','xyz'] 
soup = BeautifulSoup(sample,'html.parser')
print(soup.span['class'])
```

```python
#if we want to disable multivalues attribute and return single value instead of list
soup = BeautifulSoup(sample,'html.parser',multi_valued_attributes=None)
print(soup.span['class'])
```
