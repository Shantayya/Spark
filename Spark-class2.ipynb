```
# from pyspark import SparkConf, SparkContext
# conf = SparkConf().setAppName("Class2")
# sc = SparkContext().getOrCreate(conf=conf)

rdd = sc.textFile("/FileStore/tables/filterspark.txt")
```

```
#filter: this function removes the data from rdd. and return new rdd. 
# filter function return data by removing filtered data from rdd based on true or false condition. 
```

```
def wordFilter(x):
    if not (x.startswith('a') or x.startswith('c')):
        return True
    else:
        return False
Flatrdd = rdd.flatMap(lambda x: x.split(" "))
filterrdd5 = Flatrdd.filter(wordFilter)
filterrdd5.collect()
```

```
flatrdd = rdd.flatMap(lambda x: x.split(' '))
rdd4 = flatrdd.filter(lambda x: x.startswith('a') or x.startswith('c'))
rdd4.collect()
```

```
#Note: Its not necessary to write commands in separate line. you can write command in single line preceding with dot.
rdd.flatMap(lambda x: x.split(" ")).filter(lambda x: not (x.startswith('a') or x.startswith('c'))).collect()
```

```
#distinct(): this function fetch the distinct records from rdd.
rdd = sc.textFile("/FileStore/tables/distinctspark.txt")
rdd.flatMap(lambda x: x.split(' ')).distinct().collect()
```

```
#groupByKey(): this function used to create groups based on keys in rdd. data must be in the form of tuple.
# groupByKey() returns the iterable records.
#mapValues(list) is used to get the group data.
rdd = sc.textFile("/FileStore/tables/filterspark.txt")
fltrdd = rdd.flatMap(lambda x:x.split(" "))
tuplerdd = fltrdd.map(lambda x: (len(x),x))
grprdd = tuplerdd.groupByKey()
grprdd.collect()
```

```
rdd.flatMap(lambda x:x.split(" ")).map(lambda x: (len(x),x)).groupByKey().mapValues(list).collect()
```

```
#reducebyKey(): this function combines the data based on keys. data must be in tuple form. it always accepts paramenter to combine list based on logic.
#example:
#('this',2)
#('this',3)
#('this',5)
#('this',6)
#reduceByKey(lambda x,y:x+y)
#========>         2,3: 5
#=================>5,5: 10
#=================>10,6:16

#                  ('this',16)                ---reduction:  groupBykey(): 
#                  ('this',[2,3,5,6])         ---grouping: groupBykey(): 

rdd = sc.textFile('/FileStore/tables/reducerdd.txt')
flatrdd1 = rdd.flatMap(lambda x:x.split(' '))
maprdd1 = flatrdd1.map(lambda x:(x,len(x)))
reducerdd = maprdd1.reduceByKey(lambda x,y:x+y)
reducerdd.collect()
```

```
rdd.flatMap(lambda x:x.split(" ")).map(lambda x: (x,len(x))).groupByKey().mapValues(list).collect()
```

```
#assignment: read any file and calculate the word count.
rdd.flatMap(lambda x:x.split(' ')).filter(lambda x: len(x) != 0).map(lambda x:(x,1)).reduceByKey(lambda x,y:x+y).collect()
```

```
#finding average rating
rdd = sc.textFile('/FileStore/tables/average_quize.csv')
rdd.map(lambda x:(x.split(',')[0],(float(x.split(',')[2]),1))).reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1])).map(lambda x:(x[0],x[1][0]//x[1][1])).collect()
```

```
#finding min and max of rating using reduceByKey() function
rdd.map(lambda x:(x.split(',')[1],x.split(',')[2])).reduceByKey(lambda x,y:x if x < y else y).collect()
```
