```
#Perform the following analytics on data
#1. Show the number of students in file
#2. Show the total marks achived by females and male students.
#3. Show the total number of students passed and failed. >50 --> passed <50 -- > fail
#4. Show the total number of students enrolled per course.
#5. Show the total marks students have achived per course.
#6. Show the avg marks students have achived per course. 
#7. Show the min and max marks achived by students per course. 
#8. Show the average age of make and female students. 
```

```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName('Mini-Project')
sc = SparkContext.getOrCreate(conf = conf)
rdd = sc.textFile('/FileStore/tables/studentDatat.csv')
headers = viewrdd.first()
viewrdd = rdd.filter(lambda x: x != headers)
viewrdd.collect()
```

```
#1. Show the number of students in file
viewrdd.count()
```

```
#2. Show the total marks achived by females and male students.
viewrdd.filter(lambda x: x.split(',')[1] != 'other').map(lambda x:(x.split(',')[1],int(x.split(',')[5]))).reduceByKey(lambda x,y:x+y).collect()
```

```
#3. Show the total number of students passed and failed. >50 --> passed <50 -- > fail
viewrdd.map(lambda x:int(x.split(',')[5])).filter(lambda x: x if x > 50 else 0).count()
```

```
# fail
viewrdd.map(lambda x:int(x.split(',')[5])).filter(lambda x: not (x if x > 50 else 0)).count()
```

```
#4. Show the total number of students enrolled per course.
viewrdd.map(lambda x:(x.split(",")[3],1)).reduceByKey(lambda x,y:x+y).collect()
```

```
#5. Show the total marks students have achived per course.
viewrdd.map(lambda x:(x.split(",")[3],int(x.split(',')[5]))).reduceByKey(lambda x,y:x+y).collect()
```

```
#6. Show the avg marks students have achived per course. 
viewrdd.map(lambda x:(x.split(",")[3],(int(x.split(',')[5]),1))).reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1])).map(lambda x:(x[0],x[1][0]//x[1][1])).collect()
```

```
#7. Show the min and max marks achived by students per course. 
viewrdd.map(lambda x:(x.split(",")[3],int(x.split(',')[5]))).reduceByKey(lambda x,y:x if x > y else y).collect()
```

```
#7. Show the min marks achived by students per course. 
viewrdd.map(lambda x:(x.split(",")[3],int(x.split(',')[5]))).reduceByKey(lambda x,y:x if x < y else y).collect()
```

```
#8. Show the average age of male and female students. 
viewrdd.filter(lambda x:x.split(",")[1] != 'other').map(lambda x:(x.split(",")[1],(int(x.split(",")[0]),1))).reduceByKey(lambda x,y:(x[0]+y[0],x[1]+y[1])).map(lambda x:(x[0],x[1][0]//x[1][1])).collect()
```
