```
rdd = sc. textFile("/FileStore/tables/Sparkmap_assignment1.txt")
rdd3 = rdd.map(LengthCal)
rdd3.collect()
```

```
def LengthCal(x):
    l = x.split(" ")
    l2 = []
    for i in l:
        l2.append(len(i))
    return l2
```

```
rdd4 = rdd.map(lambda x: [len(i) for i in x.split(' ')])
rdd4.collect()
```

```
# flatmap: this function work like a map except it unifies the output in single list rather than nested list. 
# Map: each row or line in text file will act as list. so all lines will make nexted list.
```

```
maprdd = rdd.map(lambda x: x.split(" "))
maprdd.collect()
```

```
flatmaprdd = rdd.flatMap(lambda x: x.split(" "))
flatmaprdd.collect()
```
